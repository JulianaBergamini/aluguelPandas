## Criando Novas Variáveis

```python
import pandas as pd

dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
# criando uma nova variável: valor bruto (valor + condomínio + iptu)

dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
# criando uma nova variável: valor m2 (valor por metro quadrado)

dados['Valor m2'] = dados['Valor']/dados['Area']
```

```python
# arredondando os números: método round()

dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
# criando uma nova variável: valor bruto do metro quadrado

dados['Valor Bruto m2'] = (dados['Valor Bruto']/dados['Area']).round(2)
```

```python
# criando uma nova variável: tipo (casas e apartamentos)

casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# criando uma nova variável: tipo agregado

dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')

dados
```

## Excluindo variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux.head(10)
```

```python
# para excluir uma variável: método del

del dados_aux['Valor Bruto']
```

```python
dados_aux
```

```python
# para excluir uma variável: método pop()

dados_aux.pop('Valor Bruto m2')
```

```python
# para excluir múltiplas variáveis: drop()
# é necessario especificar o eixo pelo axis =
# é necessário assinalar inplace como True


dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
```

```python
dados
```

```python
# exportar e sobrescrever o arquivo:

dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
