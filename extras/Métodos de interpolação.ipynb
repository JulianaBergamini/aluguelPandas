```python
# Métodos de interpolação

import pandas as pd 

data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
data
```

```python
# fillna(): atribui valores aos campos nulos

s.fillna(0)
```

```python
# ffill(): outro método para realizar a interpolação
# analisa a Series do primeiro elemento ao último e quando encontrar um valor nulo, ele coleta o valor da última assinatura válida

s.fillna(method = 'ffill')
```

```python
# bfill(): outro método para realizar a interpolação
# realiza a mesma análise das assinaturas em uma Series, mas dessa vez de baixo para cima

s.fillna(method = 'bfill')
```

```python
# podemos também nao utilizar nenhum método para realizar a interpolação: fillna()
# o fillna() coleta a média de todos os valores não-nulos e a usa como preenchimento

s.fillna(s.mean())
```

```python
# Parâmetros para ajudar a análise quando os métodos não são suficientes:
# limit:

s.fillna(method = 'ffill', limit = 1)
```

```python
s.fillna(method = 'bfill', limit = 1)
```
