```python
import pandas as pd

dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
# análises de estatísticas descritivas
# média dos valores: mean()

dados['Valor'].mean()
```

```python
# média segundo alguns tipos de variáveis

bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]

dados['Bairro'].drop_duplicates()

grupo_bairro = dados.groupby('Bairro')

for bairro, dados in grupo_bairro: 
    print(bairro)
```

```python
# valor médio com base na lista de bairros:

for bairro, dados in grupo_bairro: 
    print('{} -> {}'.format(bairro, dados.Valor.mean()))
```

```python
# coletar a média de 'Valor' e 'Condomínio'

grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

```python
grupo_bairro['Valor'].describe().round(2)

# o desvio padrao é muito alto no Bota Fogo e na Tijuca:
```

```python
# método aggragate()

grupo_bairro['Valor'].aggregate(['min', 'max', 'sum'])
```

```python
# método rename(): renomear nomenclaturas

grupo_bairro['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'})
```

```python
#figsize(): configurar o tamanho da representacao visual

%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))

fig = grupo_bairro['Valor'].mean().plot.bar(color = 'grey')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```
